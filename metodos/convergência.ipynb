A equação f(x) = x**5 - 6*x**4 - 14*x**3 + 72*x**2 + 44*x - 180 possui três raízes positivas no intervalo [0,7]

1. Determine essas raízes pelos métodos da bissecção, secante e falsa posição.
2. Compare, graficamente, a convergência dos métodos usados no item i.

Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definindo a função f(x)

```python
def f(x):
  return x**5 - 6*x**4 - 14*x**3 + 72*x**2 + 44*x - 180
```

Desenhando o gráfico da função f(x)

```python
x = np.linspace(0, 7, 1000)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)', color='blue')
plt.axhline(0, color='red', linewidth=1)
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Gráfico da função f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

Com isso, podemos observar que, as raizes estão nos intervalos:
[1,2], [2,3] e [6,7]

```python
lista_erros_1_2 = []
lista_erros_2_3 = []
lista_erros_6_7 = []
```

### Método da bissecção

```python
def bissecao(a, b, f, erro=1e-10):
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) * f(b) >= 0: o método da bisseção requer sinais opostos em a e b.")

    erros = []
    iteracoes = 0

    while (b - a) / 2 > erro:
        m = (a + b) / 2
        r = f(m)
        erros.append(abs(b - a) / 2)  # erro atual
        if f(a) * r < 0:
            b = m
        else:
            a = m
        iteracoes += 1

    raiz = (a + b) / 2
    return raiz, erros
```

1. Intervalo [1,2]

```python
raiz, erros = bissecao(1,2,f)
print(f"Raiz: {raiz}")
lista_erros_1_2.append(erros)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(erros) + 1), erros, marker='o', color='blue')
plt.xlabel('Iteração')
plt.ylabel('Erro |b - a| / 2')
plt.title('Convergência do Método da Bisseção')
plt.grid(True)

ultima_iter = len(erros)
ultimo_erro = erros[-1]
plt.scatter(ultima_iter, ultimo_erro, color='red', label='Último ponto', zorder=5)
plt.text(ultima_iter, ultimo_erro, f"{ultima_iter}", color='red', fontsize=10, ha='left', va='bottom')

plt.legend()
plt.show()
```

### Método da secante

```python
def secante(a, b, f, erro=1e-10, max_iter=1000):
    if f(a) == f(b):
        raise ValueError("f(a) e f(b) não podem ser iguais no método da secante.")

    erros = []
    iteracoes = 0

    while abs(b - a) > erro and iteracoes < max_iter:
        c = b - f(b) * (b - a) / (f(b) - f(a))

        erros.append(abs(c - b))

        a, b = b, c
        iteracoes += 1

    return b, erros
```

### Método da falsa posição

```python
def falsa_posicao(a, b, f, erro=1e-10, max_iter=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) * f(b) >= 0: o método da falsa posição requer sinais opostos em a e b.")

    erros = []
    iteracoes = 0
    c = a

    while iteracoes < max_iter:
        c_novo = b - f(b) * (b - a) / (f(b) - f(a))
        erros.append(abs(c_novo - c))

        c = c_novo
        f_c = f(c)

        if abs(f_c) < erro or abs(b - a) < erro:
            break

        if f(a) * f_c < 0:
            b = c
        else:
            a = c

        iteracoes += 1

    return c, erros
```

### Gráfico comparando os métodos

```python
import matplotlib.pyplot as plt

def plotar_erros(lista_de_erros, nomes=None):
    plt.figure(figsize=(12, 7))

    for i, erros in enumerate(lista_de_erros):
        label = nomes[i] if nomes and i < len(nomes) else f"Método {i+1}"
        x = list(range(1, len(erros) + 1))

        (linha_plotada,) = plt.plot(x, erros, marker='o', label=label)
        cor = linha_plotada.get_color()

        ultimo_x = x[-1]
        ultimo_y = erros[-1]
        plt.scatter(ultimo_x, ultimo_y, color=cor, zorder=5)
        plt.text(ultimo_x, ultimo_y + 0.01, f"{ultimo_x}", color=cor, fontsize=13,
                 ha='left', va='bottom')

    plt.xlabel("Iteração")
    plt.ylabel("Erro")
    plt.title("Convergência dos Métodos")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plotar_erros(lista_erros_1_2, nomes=["Bisseção", "Secante", "Falsa Posição"])
```

```python
plotar_erros(lista_erros_2_3, nomes=["Bisseção", "Secante", "Falsa Posição"])
```

```python
plotar_erros(lista_erros_6_7, nomes=["Bisseção", "Secante", "Falsa Posição"])
```
